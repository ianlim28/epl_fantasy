```python
#start building the dataframe from the scraped list
football_df = pd.DataFrame(rows, columns=['player'])
new_lst = []
for col in football_df['player']:
    try:
        new_lst.append(re.search(r'\d+\.\d+',col).group())
    except:
        new_lst.append(col)
        
# creating an intermediate stage for cleaning up        
football_df['intermediate'] = new_lst
```

```python
# getting player cost
cost = []
for row in football_df['intermediate']:
    try:
        cost.append(re.search(r'\d+\.\d',row).group())
    except:
        cost.append(row)
        
# getting points scored last season
points = []
for row in football_df['intermediate']:
    try:
        points.append(re.sub(r'.*\.\d','',row))
    except:
        points.append(row)
```

```python
football_df['cost'] = cost
football_df['points'] = points

# checking data cleanliness
football_df[football_df['player'].str.contains('Sterling')]
```

```python
position = []
name_tmp = []
for row in football_df['player']:
    position.append(re.search(r'[^0-9]+',row.split('.')[0]).group()[-3:])
    name_tmp.append(re.search(r'[^0-9]+',row.split('.')[0]).group()[:-3])
```

```python
# getting name and team
name = []
team = []
for ele in name_tmp:
    team.append(ele[-3:])
    name.append(ele[:-3])
football_df['name'] = name
football_df['team'] = team
football_df['position'] = position

# double checking 
football_df[football_df['name'] == 'Salah']
```

```python
idx = football_df[(football_df['team'] == 'ers')|(football_df['team'] == 'rds')].index
final_df = football_df.drop(index = idx)
final_df = final_df[['name','position','team','cost','points']]
final_df[final_df['name'] == 'Sterling']
```

```python
# export out a copy of csv
final_df.to_csv('epl_players.csv')
```

```python
final_df['cost'] = final_df.cost.astype('float')
final_df['points'] = final_df.points.astype('float')
```

```python
# commence simplex algorithm
# source code guidance https://github.com/breeko/Fantasy_LP/blob/master/fantasy_lp_final.ipynb
salaries = {}
points = {}
for pos in final_df.position.unique():
    available_pos = final_df[final_df.position == pos]
    salary = list(available_pos[['name','cost']].set_index('name').to_dict().values())[0]
    point = list(available_pos[['name','points']].set_index('name').to_dict().values())[0]
    salaries[pos] = salary
    points[pos] = point
    
pos_num_available = {
    "GKP": 2,
    "DEF": 5,
    "MID": 5,
    "FWD": 3
}
```

```python
# checking out goalkeeper's salary
salaries['MID']
```

```python
from itertools import permutations
from pulp import *

SALARY_CAP=100
_vars = {k: LpVariable.dict(k,v,cat="Binary") for k, v, in points.items()}
```

```python
prob = LpProblem("Fantasy", LpMaximize)
rewards = []
costs = []
position_constraints = []

# Setting up the reward
for k, v in _vars.items():
    costs += lpSum([salaries[k][i] * _vars[k][i] for i in v])
    rewards += lpSum([points[k][i] * _vars[k][i] for i in v])
    prob += lpSum([_vars[k][i] for i in v]) == pos_num_available[k]
    
prob += lpSum(rewards)
prob += lpSum(costs) <= SALARY_CAP
```

```python
prob.solve()
```

```python
def summary(prob):
    div = '---------------------------------------\n'
    print("Variables:\n")
    score = str(prob.objective)
    constraints = [str(const) for const in prob.constraints.values()]
    for v in prob.variables():
        score = score.replace(v.name, str(v.varValue))
        constraints = [const.replace(v.name, str(v.varValue)) for const in constraints]
        if v.varValue != 0:
            print(v.name, "=", v.varValue)
    print(div)
    print("Constraints:")
    for constraint in constraints:
        constraint_pretty = " + ".join(re.findall("[0-9\.]*\*1.0", constraint))
        if constraint_pretty != "":
            print("{} = {}".format(constraint_pretty, eval(constraint_pretty)))
    print(div)
    print("Score:")
    score_pretty = " + ".join(re.findall("[0-9\.]+\*1.0", score))
    print("{} = {}".format(score_pretty, eval(score_pretty)))
```

```python
summary(prob)
```
